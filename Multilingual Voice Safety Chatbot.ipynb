```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import gradio as gr

# Load dataset
df = pd.read_csv("D:\Aditya\Github\Domestic violence\domestic_violence_extended_data.csv")
df.head()
```

```python
print(df.info())
print(df.describe())
print(df["State"].value_counts())
```

```python
state_totals = df.groupby("State")["Cases_Reported"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=state_totals.values, y=state_totals.index, palette="Reds_r")
plt.title("Total Domestic Violence Cases by State (20202021)")
plt.xlabel("Cases Reported")
plt.ylabel("State")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
pivot_age = df.pivot_table(index='State', columns='Age_Group', values='Cases_Reported', aggfunc='sum')

plt.figure(figsize=(10,8))
sns.heatmap(pivot_age, annot=True, cmap="YlOrBr", fmt=".0f")
plt.title("Domestic Violence by Age Group & State")
plt.show()
```

```python
yearly = df.groupby("Year")["Cases_Reported"].sum()

plt.figure(figsize=(6,4))
yearly.plot(marker='o', color='crimson')
plt.title("Year-wise Domestic Violence Cases")
plt.ylabel("Cases Reported")
plt.grid(True)
plt.show()
```

```python
urban_rural = df.groupby("Urban_Rural")["Cases_Reported"].sum()

plt.figure(figsize=(6,4))
urban_rural.plot(kind="pie", autopct="%.1f%%", colors=["#ff9999", "#66b3ff"])
plt.title("Urban vs Rural Distribution")
plt.ylabel("")
plt.show()
```

```python
marital = df.groupby("Marital_Status")["Cases_Reported"].sum()

plt.figure(figsize=(6,4))
sns.barplot(x=marital.index, y=marital.values, palette="Set2")
plt.title("Cases by Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Cases Reported")
plt.grid(True)
plt.show()
```

```python
import gradio as gr
import speech_recognition as sr
from deep_translator import GoogleTranslator
from gtts import gTTS
import tempfile

def transcribe_audio(audio):
    recognizer = sr.Recognizer()
    try:
        with sr.AudioFile(audio) as source:
            audio_data = recognizer.record(source)
        text = recognizer.recognize_google(audio_data)
        return text
    except sr.UnknownValueError:
        return " Could not understand audio."
    except sr.RequestError:
        return " Speech recognition service error."

def support_bot(message):
    try:
        message_en = GoogleTranslator(source='auto', target='en').translate(message)
    except:
        return " Translation error. Please try again.", None

    message_en = message_en.lower()

    # Basic NLP with fuzzy keyword check
    if any(word in message_en for word in ["help", "support", "assist", "emergency"]):
        response_en = " Please call 181 (Women Helpline) or 100 (Police)."
    elif "safety" in message_en:
        response_en = " Visit: https://www.ncw.nic.in/ for safe shelters and legal resources."
    elif "law" in message_en or "legal" in message_en:
        response_en = " Section 498A IPC protects women. Free legal aid is available at your local court."
    else:
        response_en = " Type or say 'help', 'safety', or 'law' for assistance."

    try:
        user_lang = GoogleTranslator(source='auto', target='en').translate(message, return_all=True)['language']
        if user_lang != "en":
            translated_response = GoogleTranslator(source='en', target=user_lang).translate(response_en)
        else:
            translated_response = response_en
    except:
        translated_response = response_en

    # Generate audio using gTTS
    try:
        tts = gTTS(translated_response)
        with tempfile.NamedTemporaryFile(delete=False, suffix=".mp3") as fp:
            tts.save(fp.name)
            return translated_response, fp.name
    except:
        return translated_response, None

with gr.Blocks(title="Multilingual Domestic Safety Voice Bot") as demo:
    gr.Markdown("##  Multilingual Domestic Safety Voice Bot")
    gr.Markdown(" Upload or record audio OR type your query:")

    with gr.Row():
        audio_input = gr.Audio(type="filepath", label=" Record or Upload Audio")
        transcribed_text = gr.Textbox(label=" Transcribed Text")
        transcribe_btn = gr.Button(" Convert Speech to Text")

    transcribe_btn.click(fn=transcribe_audio, inputs=audio_input, outputs=transcribed_text)

    with gr.Row():
        user_input = gr.Textbox(label=" Your Message")
        bot_response = gr.Textbox(label=" Bot Response")
        bot_audio = gr.Audio(label=" Bot Speech")

    # Auto-fill typed message from transcribed speech
    transcribed_text.change(fn=lambda x: x, inputs=transcribed_text, outputs=user_input)

    # Process message and return bot response + audio
    user_input.change(fn=support_bot, inputs=user_input, outputs=[bot_response, bot_audio])

demo.launch()
```
